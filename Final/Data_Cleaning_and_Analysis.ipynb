```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
# Define the directory where your CSV files are located
data_dir = '/Users/Abdullah/Desktop/All csvs'  # Replace with your actual path

# List all CSV files in the directory
csv_files = [file for file in os.listdir(data_dir) if file.endswith('.csv')]

# Initialize an empty list to hold DataFrames
df_list = []

# Loop through the CSV files and read them into DataFrames
for file in csv_files:
    file_path = os.path.join(data_dir, file)
    try:
        df = pd.read_csv(file_path, low_memory=False)
        df_list.append(df)
        print(f"Successfully loaded {file}")
    except Exception as e:
        print(f"Error loading {file}: {e}")

# Combine all DataFrames into one
combined_df = pd.concat(df_list, ignore_index=True)

print(f"Combined DataFrame shape: {combined_df.shape}")
combined_df.head()
```

```python
combined_df.to_csv("/Users/Abdullah/Desktop/combinedPath5743.csv")
```

```python
combined_df.info()
```

```python
# Specify the columns relevant to your analysis
key_columns = ['Latitude(deg)', 'Longitude(deg)', 'Time(EDT (UTC - 4 hr))']

# Drop rows where any of the key columns have NaN values
cleaned_df = combined_df.dropna(subset=key_columns).copy()

print(f"DataFrame shape after dropping NaNs: {cleaned_df.shape}")
```

```python
# Remove duplicate rows based on all columns
cleaned_df = cleaned_df.drop_duplicates()

print(f"DataFrame shape after dropping duplicates: {cleaned_df.shape}")
```

```python
# Define the datetime format
datetime_format = "%m/%d/%Y %I:%M.%S.%f %p"

# Parse the 'Time' column
cleaned_df['Time'] = pd.to_datetime(cleaned_df['Time(EDT (UTC - 4 hr))'], format=datetime_format, errors='coerce')

# Drop rows where datetime conversion failed (if any)
cleaned_df = cleaned_df.dropna(subset=['Time'])

print(f"DataFrame shape after datetime conversion: {cleaned_df.shape}")
```

```python
# Set 'Time' as the index
cleaned_df.set_index('Time', inplace=True)
```

```python
cleaned_df.info()
cleaned_df.head()
```

```python
cleaned_df.to_csv("/Users/Abdullah/Desktop/cleanedPath5743.csv")
```

```python
# Resample to every 15 minutes, keeping the first available record in each interval
filtered_df = cleaned_df.resample('15T').first()

print(f"Filtered DataFrame shape: {filtered_df.shape}")
filtered_df.head()
```

```python
filtered_df.to_csv("/Users/Abdullah/Desktop/15minPath5743.csv")
```

```python
# Reset index to access the time column
filtered_df_reset = filtered_df.reset_index()

# Convert 'Time' to Unix timestamp (seconds since epoch)
filtered_df_reset['Timestamp'] = filtered_df_reset['Time'].astype(int) / 10**9
filtered_df_reset.head()
```

```python
# Compute correlation matrix
correlation_matrix_15min = filtered_df_reset[['Latitude(deg)', 'Longitude(deg)', 'Timestamp']].corr()

print("Correlation Matrix (15-Minute Intervals):")
print(correlation_matrix_15min)
```

```python
# Visualize correlation matrix with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix_15min, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix (15-Minute Intervals)')
plt.show()
```
